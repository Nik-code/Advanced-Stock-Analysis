```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
# Check contents of the data directory
data_path = '../data'
```

```python
# Function to load all CSV files from a directory
def load_csv_files(directory):
    dataframes = {}
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            df = pd.read_csv(file_path)
            stock_code = filename.split('.')[0]  # Assuming filenames are like '500002.csv'
            dataframes[stock_code] = df
    return dataframes
```

```python
# Load all CSV files
all_data = load_csv_files(data_path)

print(f"Total number of stocks loaded: {len(all_data)}")
```

```python
# Display basic info for the first few dataframes
for i, (stock_code, df) in enumerate(all_data.items()):
    print(f"\nStock Code: {stock_code}")
    print(df.info())
    print("\n" + "="*50)
    if i == 4:  # Only show the first 5 stocks
        break
```

```python
# Analyze the first dataframe in detail
first_stock_code = next(iter(all_data))
df = all_data[first_stock_code]

print(f"\nDetailed analysis for Stock Code: {first_stock_code}")
display(df.head())
display(df.describe())

# Check for missing values
print("\nMissing Values:")
display(df.isnull().sum())
```

```python
# Create a correlation heatmap for numerical columns
plt.figure(figsize=(12, 10))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title(f'Correlation Heatmap of Numerical Features for Stock {first_stock_code}')
plt.show()
```

```python
# Visualize the distribution of closing prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Close Price'], kde=True)
plt.title(f'Distribution of Closing Prices for Stock {first_stock_code}')
plt.show()
```

```python
# Plot the closing price over time
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(df['Date']), df['Close Price'])
plt.title(f'Closing Price Over Time for Stock {first_stock_code}')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate daily returns
df['Daily Return'] = df['Close Price'].pct_change()

# Plot the distribution of daily returns
plt.figure(figsize=(10, 6))
sns.histplot(df['Daily Return'].dropna(), kde=True)
plt.title(f'Distribution of Daily Returns for Stock {first_stock_code}')
plt.show()
```

```python
# Calculate and plot moving averages
df['MA50'] = df['Close Price'].rolling(window=50).mean()
df['MA200'] = df['Close Price'].rolling(window=200).mean()

plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(df['Date']), df['Close Price'], label='Close Price')
plt.plot(pd.to_datetime(df['Date']), df['MA50'], label='50-day MA')
plt.plot(pd.to_datetime(df['Date']), df['MA200'], label='200-day MA')
plt.title(f'Stock Price and Moving Averages for Stock {first_stock_code}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
